# Visualization

### What is the distribution of the 'salary' column?

```python
plt.figure(figsize=(8, 6))
plt.hist(data['salary'], bins=20, color='skyblue')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

### How does experience level impact salaries in data science?

```python
plt.figure(figsize=(5,5))
sns.barplot(data=data, x='experience_level', y='salary' , ci= None)
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.title('Salary vs Experience Level')
plt.show()
```

### What are the different company locations and their employee counts?

```python
# Count of employees in each company location
location_counts = data['company_location'].value_counts().nlargest(10)
print("Employees per company location:")
print(location_counts)
```

### What is the average salary for each job title?

```python
avg_salary_by_title = data.groupby('job_title')['salary'].mean().sort_values(ascending=False).nlargest(10)
print("Average salary by job title:")
print(avg_salary_by_title)
```

### What is the average salary for each job category?

```python
avg_salary_by_category = data.groupby('job_category')['salary'].mean().sort_values(ascending=False).nlargest(10)
print("Average salary by job category:")
print(avg_salary_by_category)
```

### What are the different company sizes, and how many employees are in each size category?

```python
# Count of employees in each company size category
company_size_counts = data['company_size'].value_counts()
print("Employees per company size category:")
print(company_size_counts)
```

### Job categories by employment type

```python
job_category_employment = pd.crosstab(data['job_category'], data['employment_type'])
print("Job categories by employment type:")
print(job_category_employment)
```

```python
def salary_in_category(category,df):
    df1=df[df.job_category== category]
    salary= df1.groupby('job_title')["salary"].mean()
    return salary
```
