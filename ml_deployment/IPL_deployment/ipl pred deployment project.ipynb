# Loading Libray

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

# Importing dataset after EDA is done 

```python
data = pd.read_csv(r"C:\Users\HP\Documents\final_df_after_eda_IPL_pred.csv")
data
```

# Spiliting x and y

```python
# Sepratating & assigning features and target columns to x & y respectively

y = data['total']
x = data.drop('total',axis=1)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=42)
```

# Train model with XGBRegressor because its gives highest accuracy

```python
xgb = XGBRegressor()
```

```python
xgb.fit(x_train, y_train)
```

```python
y_pred = xgb.predict(x_train)
Train_Accuracy = r2_score(y_pred, y_train)
Train_Accuracy
```

```python
print('Accuracy score of the training data : ', round(Train_Accuracy*100,4),"%")
```

```python
y_pred_test = xgb.predict(x_test)
Test_Accuracy = r2_score(y_pred_test , y_test)
Test_Accuracy
```

```python
print('Accuracy score of the test data : ', round(Test_Accuracy*100,4),"%")
```
