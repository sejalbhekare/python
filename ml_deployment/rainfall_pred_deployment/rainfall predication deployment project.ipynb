# importing library

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Load dataframe after EDA done

```python
# loading the rainfall predication dataset to a pandas DataFrame
rainfall_pred_dataset = pd.read_csv(r'"C:\Users\HP\Documents\df_after_eda_rainfall.csv"')
rainfall_pred_dataset
```

```python
# number of rows and Columns in this dataset
rainfall_pred_dataset.shape
```

```python
# getting the statistical measures of the data
rainfall_pred_dataset.describe()
```

```python
rainfall_pred_dataset['RainTomorrow'].value_counts()
```

# Separated the data into X and Y

```python
x = rainfall_pred_dataset.drop('RainTomorrow', axis=1)
y = rainfall_pred_dataset['RainTomorrow']
```

# Since Target Column Imbalance, we do oversampling 

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x, y)
```

```python
x=x_ros
y=y_ros
```

```python
y.value_counts()
```

```python
print(x)
```

```python
print(y)
```

# Train Test Split of the Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.3, stratify=y, random_state=42)
```

# Training the Model on random forest ML - Algorithm

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, Y_train)
```

# rf: Model Evaluation

```python
Y_pred = rf.predict(X_train)
Train_Accuracy = accuracy_score(Y_pred, Y_train)
```

```python
print('Accuracy score of the training data : ', round(Train_Accuracy*100,2),"%")
```

```python
# accuracy score on the test data
Y_pred_on_test_data = rf.predict(X_test)
Test_Accuracy = accuracy_score(Y_pred_on_test_data, Y_test)
```

```python
print('Accuracy score of the test data : ', round(Test_Accuracy*100,2),"%")
```

# Making a Predictive System

```python
input_data = [1.1134,1.4871,-0.2789,1.1975,2.2845,-0.9953,-1.1437,-0.7343,-9.6654,-1.9477,-7.1419,1231,32,0,0]

# changing the input_data list into numpy array
input_data_as_numpy_array = np.array(input_data)

# reshape the array into 1 row and all columns-type,  as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

print(input_data_reshaped)
```

```python
input_data = (1.1134,1.4871,-0.2789,1.1975,2.2845,-0.9953,-1.1437,-0.7343,-9.6654,-1.9477,-7.1419,1231,32,0,0)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array into 1 row and all columns-type,  as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = rf.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('It will not rain tomorrow')
else:
  print('it may rain tomorrow')
```
